spaCy is a fast, modern Python library for NLP that processes text in one line and automatically does tokenization, POS tagging, lemmatization, named entity recognition and more — designed for real-world production use unlike NLTK which is more for learning.

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
type(nlp)
```

```python
# Reading a document
doc1 = nlp('We are learning spacy. It is used for NLP')
doc1
```

```python
# nlp()  → runs the text through the entire spaCy pipeline

# Automatically does ALL of these at once:
# → Tokenization    (splits into words)
# → POS tagging     (labels each word's grammar role)
# → Lemmatization   (finds base form of each word)
# → NER             (finds names, places, dates)
# → Dependency parsing (understands sentence structure)

# One line does everything NLTK needed 5 steps for!
```

## Lemmatization

```python
doc3 = nlp('I am playing football i have played a lot i am a good player')
for word in doc3:
    print(word.text, word.lemma_)
```

```python
lemmatized_words = [word.lemma_ for word in doc3]
lemmatized_words
```

## Named Entity Recognition

Named Entity Recognition (NER) is automatically identifying and classifying real-world named things in text — like people, places, organizations, dates and money.


```python
doc5 = nlp('By 2025, India will grow so much in Technical field and earn more than 5 billion dollars')
doc5.ents
```

```python
for word in doc5.ents:
    print(word.text, word.label_)
```

```python
spacy.explain('GPE')
```

```python
displacy.render(doc5, style = 'ent')
```

## Semantic Similarity

```python
word1 = nlp('horse')
word2 = nlp('elephant')

word1.similarity(word2)
```

```python
# Measures how SIMILAR two words are in meaning
# Returns a number between 0 and 1

# 0.0 → completely unrelated
# 0.5 → somewhat related
# 1.0 → identical meaning
```

```python
word1 = nlp('tiger')
word2 = nlp('lion')

word1.similarity(word2)
```

```python
word1 = nlp('happy')
word2 = nlp('glad')

word1.similarity(word2)
```

```python
doc6 = nlp('Happy and glad to have a horse , cat and fish')
for w1 in doc6:
    for w2 in doc6:
        print((w1.text, w2.text), 'Similarity :- ', w1.similarity(w2))
```
